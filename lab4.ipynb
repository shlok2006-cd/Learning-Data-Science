# Question 1

The mean income of the random sample of 100 employees of an industrial concern was found to be Rs.3000. If the standard deviation of the population was 25, find the standard error of the mean and also test whether the sample mean differ from the population mean of 2850. State the null and alternative hypotheses. AT 95% confidence level, is there enough evidence to support the null hypothesis?


```python
import math
from scipy.stats import norm
```

```python
sample_mean = 3000
population_mean = 2850
sigma = 25
sample_size = 100
alpha = 0.05 #confidence level = 95%
```

```python
SE = sigma / math.sqrt(sample_size)
print("Standard Error",SE)
```

```python
Z_Calculated = (sample_mean - population_mean) / SE
print("Z Score",Z_Calculated)
```

```python
Z_Standard = norm.ppf(1-alpha/2)
print("Standard Z-Value : ",Z_Standard)
```

```python
if abs(Z_Calculated)>Z_Standard:
    print("Reject H0:Sample mean differs significantly from the population mean")
else:
    print("Fail to reject H0: Not enough evidence to say the means differ")
```

# Question 2

A truckling firm is suspicious of the claim that the average lifetime of a certain tires is at least 28000 miles. To check the claim, the firm puts 40 of these tires on its trucks and gets a mean life time of 27436 miles with a standard deviation of 1348 miles. What can it conclude if the probability of type I errors is to be at most 0.01?

```python
sample_mean=27436
sample_size=40
sigma=1348
population_mean=28000
alpha=0.01
```

```python
SE= sigma/math.sqrt(sample_size)
print("Standard Error",SE)
```

```python
z_calculated=(sample_mean-population_mean)/SE
print("Z score",z_calculated)
```

```python
z_standard=norm.ppf(alpha)
print("Standard z-value:",z_standard)
```

```python
if abs(z_calculated>z_standard):
    print("Reject H0:Sample mean differs significantly from the population mean")
else:
    print("Fail to reject H0: Not enough evidence to say the means differ")
```

```python
#using p value

p_value=norm.cdf(z_calculated)
print("P value:",p_value)
```

```python
reject_H0=p_value<alpha
```

```python
print(f"Standard error(SE)   ={SE:.4f}")
print(f"Z-statistic          ={z_calculated:.4f}")
print(f"Critical Z(alpha={alpha}) = {z_standard:.4f}")
print(f"One tailed p-value      = {p_value:.6f}")
print()
if reject_H0:
    print(f"Decision: REJECT H0 (p {p_value:.6f}<{alpha})")
    print("Conclusion:Evidence suggests the true mean lifetime is less than 28000 miles")
else:
    print(f"Decision:FAIL TO REJECT H0 (p {p_value:.6f}>={alpha})")
    print("Conclusion: Not enough evidence to say the mean is less than 28000 miles")
```

# Question 3

A manufacture of gun powder has developed a new powder which is designed to produce a muzzle equal to 3000ft/sec. Seven shells are loaded with the charge and the muzzle velocities are measured. The resulting velocitieds are as follows: 3005,2935,2995,2965,3905,2395,2905. Do these data present sufficient evidence to indicate that the average velocity differs from 3000ft/sec (use alpha=5%)

```python
import numpy as np
from scipy import stats
```

```python
data=np.array([3005,2935,2995,2965,3905,2395,2905])
mu_0=3000
alpha=0.05
n=len(data)
mean_value=np.mean(data)
sample_std=np.std(data,ddof=1)

print("sample mean: ",mean_value)
print("sample standard deviation: ",sample_std)
print("sample size: ",n)
```

```python
calculated_T=(mean_value-mu_0)/(sample_std/np.sqrt(n))

print("Calculated t-statistic:",calculated_T)
```

```python
df=n-1

critical_t=stats.t.ppf(1-alpha/2,df)
print("Degrees of freedom:",df)
print("Critical t-value:",critical_t)
```

```python
if(calculated_T>critical_t):
    print("Null hypothesis rejected")
else:
    print("Failed to reject null hypothesis due to lack of evidence")
```
