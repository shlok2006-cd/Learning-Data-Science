```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
xpoints=np.array([0,6])
ypoints=np.array([0,250])

plt.plot(xpoints,ypoints)
#plt.show() optional can show and not as well its your wish
```

```python
xpoints=np.array([1,8])
ypoints=np.array([3,10])

plt.plot(xpoints,ypoints,'o') #Plotting without line
plt.plot(xpoints,ypoints,'*') #plotting without line
```

```python
xpoints=np.array([1,2,6,8])
ypoints=np.array([3,8,1,10])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
#if we do not specify the points on x-axis they will get default values 0,1,2,3,4,5
ypoints=np.array([3,8,1,10,5,7])
plt.plot(ypoints)
```

```python
ypoints=np.array([3,8,1,10])
#plt.plot(ypoints,marker='o') #keyword argument marker to emphasize each point with a specific symbol
plt.plot(ypoints,marker='*')
plt.show()
```

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='o',color="black") #keyword argument marker to emphasize each point with a specific symbol
#plt.plot(ypoints,marker='*')
plt.show()
```

```python
plt.plot(ypoints,':g')
plt.show() #":" means dotted lines and r means red color
'''Syntax marker|line|color'''
```

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='o',ms=30)#argument markersize or the shorter version,ms to 
```

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='o',ms=30,mec='b')
plt.plot(ypoints,marker='o',ms=30,mfc='r')
```

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='o',linewidth=20.5)
```

```python
y1=np.array([3,8,1,10])
y2=np.array([6,2,7,11])
y3=np.array([1,5,6,3])
y4=np.array([9,5,4,2])
plt.plot(y1)
plt.plot(y2)
plt.plot(y3)

#plt.grid() gives x and y grid
#plt.grid(axis='y') #gives only horizontal grid
plt.grid() #gives only vertical grid

font1={'family':'poppins','color':'blue','size':20}
font2={'family':'poppins','color':'darkred','size':15}
plt.plot(y1,marker='o',ms=15,linestyle='--')
plt.plot(y2,marker='*',ms=10,color='green')
plt.plot(y3,marker='o',ms=15,linewidth='10',mec='beige')


plt.title("Sports Watch Distribution",fontdict=font1)
plt.xlabel("Average pulse",fontdict=font2)
plt.ylabel("Calorie Burnage",fontdict=font2)
```

```python
#subplot() function

x1=np.array([0,1,2,3])
y1=np.array([3,8,1,10])

plt.subplot(1,2,1) #1 means first plot contains 1 row and 2 means contains 2 columns and 1 means it is for first plot
plt.plot(x1,y1)

x2=np.array([0,1,2,3])
y2=np.array([10,20,30,40])

plt.subplot(2,2,2)   
plt.plot(x2,y2)
```

```python
#scatter plot

x=np.random.randint(100,size=(100))
y=np.random.randint(100,size=(100))
colors=np.random.randint(100,size=(100))
sizes=10*np.random.randint(100,size=(100))

plt.scatter(x,y,c=colors,s=sizes,alpha=0.5,cmap='nipy_spectral')

plt.colorbar()
```

```python
#bar plots

x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])

plt.bar(x,y)
```

```python
#histogram plot

x=np.random.normal(170,10,250)
print(x)
```

```python
plt.hist(x,color="green")
plt.show()
```

```python
#pie chart

y=np.array([35,25,25,15])
mylabels=["Apple","Bananas","Cherries","Grapes"]
myexplode=[0.2,0,0,0]
plt.pie(y,labels=mylabels,explode=myexplode,shadow=True)
plt.show()
```
