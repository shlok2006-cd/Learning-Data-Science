# Python plotly 

Plotly is a visualization tool that allows users to plot data interactively—meaning you can zoom, pan, hover, and toggle data—making it widely used in data science, analytics, and web applications.


```python
import plotly.express as px
df=px.data.tips()
fig=px.pie(df,values="total_bill",names="day")
fig.show()
```

```python
df=px.data.iris()

fig=px.scatter_3d(df,x='sepal_width',
                  y='sepal_length',
                  z='petal_width',
                  color='species')
fig.show()
```
